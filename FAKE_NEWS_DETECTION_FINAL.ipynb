```python
df_marge = pd.concat ([df_fake, df_true], axis=0)
df_marge.head(10)
```

```python
df = df_marge. drop(["title", "subject", "date"], axis=1)
df. head(10)
```

```python
df = df. sample (frac=1)
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
def word_drop(text):
     text = text.lower()
     text = re.sub('\[. *?\]','', text)
     text = re.sub("\\W"," ",text)
     text = re.sub('https?://\S+|www\. \S+', '', text)
     text = re.sub('<.*?>+', '', text)
     text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
     text = re.sub( '\n', '',text)
     text = re.sub ('\w*\d\w*','', text)
     return text
```

```python
df["text"] = df ["text"].apply (word_drop)
```

```python
df.head(10)
```

```python
x = df["text"]
y = df["class"]
```

```python
x_train, x_test, y_tarin, y_test = train_test_split(x,y, test_size = .25)
```

```python
from sklearn. feature_extraction.text import TfidfVectorizer
```

```python
vectrorization = TfidfVectorizer ()
xv_train = vectrorization.fit_transform(x_train)
xv_test = vectrorization.transform(x_test)
```

**`Classification models`**

### **LOGISTIC REGRESSION CLASSIFIER (pre trained ml model):**

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()
LR.fit (xv_train, y_tarin)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(xv_train, y_tarin)
```

```python
LR.score(xv_test, y_test)
```

```python
pred_LR = LR.predict (xv_test)
```

** Metrics used to evaluate the performance of a classification model. **

```python
print(classification_report (y_test, pred_LR))
```
